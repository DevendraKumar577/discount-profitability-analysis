```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Sample_ Superstore.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace("-", "_")
```

```python
df.columns
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"] = pd.to_datetime(df["ship_date"], errors="coerce")
```

```python
df[["order_date", "ship_date"]].head()
```

```python
df["order_date"].isnull().sum(), df["ship_date"].isnull().sum()
```

```python
df["profit_margin"] = df["profit"] / df["sales"]
```

```python
df["discount_band"] = "30%+"

df.loc[df["discount"] <= 0.30, "discount_band"] = "20-30%"
df.loc[df["discount"] <= 0.20, "discount_band"] = "10-20%"
df.loc[df["discount"] <= 0.10, "discount_band"] = "0-10%"
```

```python
df["loss_flag"] = "Profit"
df.loc[df["profit"] < 0, "loss_flag"] = "Loss"
```

```python
df["high_discount_flag"] = "No"
df.loc[df["discount"] > 0.20, "high_discount_flag"] = "Yes"
```

```python
df[
    ["sales", "profit", "discount",
     "profit_margin", "discount_band", "loss_flag"]
].head()
```

```python
df.to_csv("clean_superstore.csv", index=False)
```
